```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('../Computer-Vision-with-Python/DATA/00-puppy.jpg', cv2.IMREAD_COLOR)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img_rgb)
```

```python
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
plt.imshow(img_hsv)

newImg = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2HLS)
plt.imshow(newImg)
```

```python
img_hsl = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
img_hsl2 = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2HLS)
plt.imshow(img_hsl)
plt.imshow(img_hsl2)
```

```python
img_hls = cv2.cvtColor(img_rgb, cv2.COLOR_BGR2HLS)
plt.imshow(img_hls)
```

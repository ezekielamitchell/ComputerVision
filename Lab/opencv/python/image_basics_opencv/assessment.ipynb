```python
image = cv2.imread(img2)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
vertices = np.array([[2250,3250],[2950,2250],[3500,3250]], np.int32)
vertices
vertices.shape
```

# Drawing on Images

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
blank_image = np.zeros(shape=(512,512,3), dtype=np.int16)
plt.imshow(blank_image)
```

```python
plt.imshow(blank_image)
```

```python
blank_image=np.zeros(shape=(512,512,3), dtype=np.int16)
cv.rectangle(blank_image, pt1=(150,150), pt2=(350,350), thickness=10, color=(0,0,255))
plt.imshow(blank_image)
```

```python
cv.circle(blank_image, center=(250,250), radius=100, color=(0,0,255), thickness=-1)
cv.putText(blank_image, "Endr", org=(150,150), fontFace=cv.FONT_HERSHEY_DUPLEX, fontScale=3, color=(255,255,255))
plt.imshow(blank_image)
```

# Writing on Images

```python
font = cv.FONT_HERSHEY_SIMPLEX
cv.putText(blank_image, "Endr", org=(150,450), fontFace=font, fontScale=2,
           color=(255,255,255), thickness=2, lineType=cv.LINE_AA)
plt.imshow(blank_image)
```

```python
blank_image = np.zeros((512,512,3), dtype=np.int32)
plt.imshow(blank_image)
```

```python
vertices = np.array([[100,400], [200,300], [300,400], [200,500]], dtype=np.int32)
print(vertices.shape)
pts = vertices.reshape((-1,1,2))
pts.shape
cv.polylines(blank_image, pts=[pts], isClosed=True, color=(0,255,255), thickness=3)
plt.imshow(blank_image)
```

# Image Basics with OpenCV

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_path = '/home/ezekiel/Developer/ComputerVision/Lab/opencv/data/images/image04.jpg'
image = cv.imread(image_path)
type(image)
```

```python
plt.imshow(image)
```

```python
fixed_image = cv.cvtColor(image, cv.COLOR_BGR2RGB)
```

```python
plt.imshow(fixed_image)
```

# Resizing Images

```python
fixed_image.shape
```

```python
x = int(fixed_image.shape[0] / 2)
y = int(fixed_image.shape[1] / 2)
new_size = (y, x)  # Note: OpenCV expects (width, height)
new_image = cv.resize(fixed_image, new_size)
```

```python
print(fixed_image.shape)
print(new_image.shape)
```

```python
plt.imshow(new_image)
```

# Color Spaces

In the 1970's HSL and HSV (Hue, Saturation, Lightness/Value) were developed as alternative color models
HSL/V are more closely alined to how humans view color

```python
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('../data/image04.jpg') # import image via opencv
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB) # change to correct color mapping to RGB
plt.imshow(image) # display image via matplotlib
```

```python
image = cv2.imread('../data/image04.jpg') # import image via opencv
image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV) # change to correct color mapping to HSV
plt.imshow(image) # display image via matplotlib
```

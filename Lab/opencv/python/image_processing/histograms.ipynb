```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def load_image():
    blank_image = np.zeros((600,600))
    font = cv2.FONT_HERSHEY_DUPLEX
    cv2.putText(blank_image, 'ENDR', (50,300), font, 6, (255,255,255), 20)
    return blank_image

def display_image(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap="gray")
```

```python
dark_horse = cv2.imread('../data/horse.jpg')
show_horse = cv2.imread('../data/horse.jpg', cv2.COLOR_BGR2RGB)

rainbow = cv2.imread('../data/rainbow.jpg')
show_rainbow = cv2.imread('../data/rainbow.jpg', cv2.COLOR_BGR2RGB)

blue_bricks = cv2.imread('../data/bricks.jpg')
show_bricks = cv2.imread('../data/bricks.jpg', cv2.COLOR_BGR2RGB)
```

```python
hist_values_bricks = cv2.calcHist([blue_bricks], channels=[0], mask=None, histSize=[256], ranges=[0,256])
hist_values_rainbow = cv2.calcHist([rainbow], channels=[0], mask=None, histSize=[256], ranges=[0,256])
hist_values_horse = cv2.calcHist([dark_horse], channels=[0], mask=None, histSize=[256], ranges=[0,256])
```

```python
plt.plot(hist_values_bricks)
```

```python
plt.plot(hist_values_rainbow)
```

```python
plt.plot(hist_values_horse)
```

```python
img = blue_bricks
color = ('b', 'g', 'r')

for i, col in enumerate(color):
    histr = cv2.calcHist([img], [i], None, [256], [0,256])
    plt.plot(histr, color=col)
    plt.xlim([0,256])

plt.title("Histogram for Blue Bricks")
```

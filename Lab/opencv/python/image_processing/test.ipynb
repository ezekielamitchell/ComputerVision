```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def load_image():
    blank_image = np.zeros((600,600))
    font = cv2.FONT_HERSHEY_DUPLEX
    cv2.putText(blank_image, 'ENDR', (50,300), font, 6, (255,255,255), 20)
    return blank_image

def display_image(img, cmap=None):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap)

```

```python
img = cv2.imread("../data/giraffes.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
display_image(img)
```

```python
img = cv2.imread("../data/giraffes.jpg", 0)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
ret, th1 = cv2.threshold(img, 120,255,cv2.THRESH_BINARY)
```

```python
display_image(th1)
```

```python
img = cv2.imread("../data/giraffes.jpg")
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
plt.imshow(img_hsv)
```

```python
img = cv2.imread("../data/giraffes.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
low_pass = np.ones((4,4), np.float32) * 1/10
low_pass
```

```python
img_new = cv2.filter2D(img, -1, low_pass)
display_image(img_new)
```

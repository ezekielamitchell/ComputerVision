# Opening an Image using OpenCV

```python
import cv2 as cv
from time import sleep
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load an image
data_set = '/home/ezekiel/Developer/ComputerVision/Lab/opencv/data/images/'
image = cv.imread(data_set + 'image01.jpg')
type(image)
```

```python
if image is None:
    print('Error, image not found')
else:
    print('Loading image...')
    sleep(1.5)
    cv.imshow('Image', image)

    cv.waitKey(0)
    print('Destroying windows...')
    sleep(1.5)
    cv.destroyAllWindows()
```

```python
from PIL import Image
```

```python
image = Image.open(data_set + 'image01.jpg')
type(image)

image_arr = np.asarray(image)
plt.imshow(image)
pic_red = image_arr.copy()
pic_blue  = image_arr.copy()
pic_green = image_arr.copy()
```

```python
type(pic_red)
```

```python
plt.imshow(pic_red[:,:,0], cmap='gray') # red grayscale
```

```python
plt.imshow(pic_blue[:,:,2], cmap='gray') # green grayscale 
```

```python
plt.imshow(pic_green[:,:,1], cmap='gray') # blue grayscale
```

```python
pic_red[:,:,1]=0
pic_green[:,:,2]=0
pic_blue[:,:,0] = 0
pic_blue[:,:,1] = 0
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('../Computer-Vision-with-Python/DATA/sammy_face.jpg')
plt.imshow(img)
```

```python
edges = cv2.Canny(image=img,threshold1=220,threshold2=220)
```

```python
plt.imshow(edges)
```

```python
# median pixel value
med_val = np.median(img)
med_val
```

```python
# set to 0 or 70% of median (whichever greater)
lower_th = int(max(0,0.7*med_val))
# set to max or 130% of median (whichever smaller)
upper_th = int(min(255,1.3*med_val))
```

```python
edges = cv2.Canny(img,lower_th,upper_th)
```

```python
plt.imshow(edges)
```

```python
blurr = cv2.blur(img, (5,5))
```

```python
edges = cv2.Canny(blurr,lower_th+50,upper_th+50)
```

```python
plt.imshow(edges)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
denis = cv2.imread('../Computer-Vision-with-Python/DATA/Denis_Mukwege.jpg',0)
nadia = cv2.imread('../Computer-Vision-with-Python/DATA/Nadia_Murad.jpg',0)
solvay = cv2.imread('../Computer-Vision-with-Python/DATA/solvay_conference.jpg',0)
```

```python
plt.imshow(solvay,cmap='gray')
```

```python
face_cascade = cv2.CascadeClassifier('../Computer-Vision-with-Python/DATA/haarcascades/haarcascade_frontalface_default.xml')
```

```python
def detect_face(img):
    face_img = img.copy()
    face_rectangles = face_cascade.detectMultiScale(face_img)

    for (x,y,w,h) in face_rectangles:
        cv2.rectangle(face_img,(x,y),(x+w,y+h),(255,255,255),10)

    
    plt.imshow(face_img,cmap='gray')
    return face_img
```

```python
detect_face(nadia)
```

```python
def adjusted_detect_face(img):
    face_img = img.copy()
    face_rectangles = face_cascade.detectMultiScale(face_img,scaleFactor=1.1,minNeighbors=5)

    for (x,y,w,h) in face_rectangles:
        cv2.rectangle(face_img,(x,y),(x+w,y+h),(255,255,255),10)

    
    plt.imshow(face_img,cmap='gray')
    return face_img
```

```python
adjusted_detect_face(solvay)
```

```python
eye_cascade = cv2.CascadeClassifier('../Computer-Vision-with-Python/DATA/haarcascades/haarcascade_eye.xml')
def detect_eyes(img):
    face_img = img.copy()
    face_rectangles = eye_cascade.detectMultiScale(face_img,scaleFactor=1.2,minNeighbors=5)

    for (x,y,w,h) in face_rectangles:
        cv2.rectangle(face_img,(x,y),(x+w,y+h),(255,255,255),10)

    
    plt.imshow(face_img,cmap='gray')
    return face_img
```

```python
detect_eyes(nadia)
```

```python
cap = cv2.VideoCapture(0)
```

```python
while True:
    ret, frame = cap.read(0)
    frame = detect_face(frame)
    cv2.imshow('Face Detection',frame)

    if cv2.waitKey(1) == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
    
```

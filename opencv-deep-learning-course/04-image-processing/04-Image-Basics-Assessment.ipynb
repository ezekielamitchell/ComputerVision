#### TASK: Open the *dog_backpack.jpg* image from the DATA folder and display it in the notebook. Make sure to correct for the RGB order.

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv.imread('../data/dog_backpack.jpg')
img_fix = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img_fix)
```

#### TASK: Flip the image upside down and display it in the notebook.

```python
img = img_fix.copy()
plt.imshow(cv.flip(img, 0))
```

#### TASK: Draw an empty RED rectangle around the dogs face and display the image in the notebook.

```python
img = img_fix.copy()
plt.imshow(img)
```

```python
img_new = img.copy()
pt1, pt2 = (200,400), (600,800)
cv.rectangle(img_new, pt1, pt2, color=(255,0,0), thickness=5)
plt.imshow(img_new)
```

#### TASK: Draw a BLUE TRIANGLE in the middle of the image. The size and angle is up to you, but it should be a triangle (three sides) in any orientation.

```python
img_new = img.copy()
pts = np.array([[200,750], [400,350], [600,750]], dtype=np.int32)
pts = pts.reshape(-1,1,2)
cv.polylines(img_new, [pts], isClosed=True, color=(0,0,255), thickness=5)
plt.imshow(img_new)
```

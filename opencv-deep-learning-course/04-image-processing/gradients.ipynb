```python
# Import required libraries
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def display_img(img):
    fig = plt.figure(figsize=(12,10))
    axe = fig.add_subplot(111)
    axe.imshow(img, cmap='gray')
```

```python
img = cv.imread('../data/sudoku.jpg', 0)
display_img(img)
```

```python
sobelx = cv.Sobel(img, cv.CV_64F, dx=1, dy=0, ksize=3)
```

```python
display_img(sobelx)
```

```python
laplacian = cv.Laplacian(img, cv.CV_64F)
```

```python
display_img(laplacian)
```

```python
sobely= cv.Sobel(img, cv.CV_64F, 0, 1)
blended = cv.addWeighted(src1=sobelx, alpha=0.5, src2=sobely, beta=0.5, gamma=0)
```

```python
display_img(blended)
```

```python
display_img(laplacian)
```

```python
ret, th1 = cv.threshold(img, 100, 255, cv.THRESH_BINARY)
display_img(th1)
```

```python
kernel = np.ones((4,4), np.uint8)
gradient = cv.morphologyEx(blended, cv.MORPH_GRADIENT, kernel)
```

```python
display_img(gradient)
```

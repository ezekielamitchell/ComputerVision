```python
# Import required libraries
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dark_horse = cv.imread('../data/horse.jpg')
show_horse = cv.cvtColor(dark_horse, cv.COLOR_BGR2RGB)

rainbow = cv.imread('../data/rainbow.jpg')
show_rainbow = cv.cvtColor(rainbow, cv.COLOR_BGR2RGB)

bricks = cv.imread('../data/bricks.jpg')
show_bricks = cv.cvtColor(bricks, cv.COLOR_BGR2RGB)
```

```python
def display_img(img):
    fig = plt.figure(figsize=(12,10))
    axe = fig.add_subplot(111)
    axe.imshow(img, cmap='gray')
```

```python
plt.imshow(show_bricks)
```

```python
hist1 = cv.calcHist([bricks], channels=[0], mask=None, histSize=[256], ranges=[0,256])
```

```python
hist1.shape
```

```python
plt.plot(hist1)
```

```python
hist2 = cv.calcHist([dark_horse], channels=[0], mask=None, histSize=[256], ranges=[0,256])
plt.plot(hist2)
```

```python
hist3 = cv.calcHist([rainbow], channels=[0], mask=None, histSize=[256], ranges=[0,256])
plt.plot(hist3)
```

```python
hist_merge =  hist3 + hist1
plt.plot(hist_merge)
```

```python
img = bricks
color = ( 'b', 'g', 'r' )
for i, color in enumerate(color):
    histr = cv.calcHist([img], [i], None, [256], [0,256])
    plt.plot(histr, color=color)
    plt.xlim([0,256])
    plt.title('HISTOGRAM - BRICKS')
```

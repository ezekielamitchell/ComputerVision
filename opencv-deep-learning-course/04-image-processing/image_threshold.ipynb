```python
img = cv.imread('../data/crossword.jpg', 0)
plt.imshow(img, cmap='gray')
```

```python
def show_pic(img):
    fig = plt.figure(figsize=(15,15))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap='gray')
```

```python
show_pic(img)
```

```python
ret, th1 = cv.threshold(img, 127, 255, cv.THRESH_BINARY)
show_pic(th1)
```

```python
th2 = cv.adaptiveThreshold(img, 255, cv.ADAPTIVE_THRESH_MEAN_C, cv.THRESH_BINARY, 5, 8)
show_pic(th2)
```

```python
blended = cv.addWeighted(th1, 0.6, th2, 0.4, 0)
show_pic(blended)
```

```python
# Import required libraries
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def show(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap='gray')
```

```python
def load_img():
    blank_img = np.zeros((600,600))
    font = cv.FONT_HERSHEY_SIMPLEX
    cv.putText(blank_img, text="ABCDE", org=(50,300), fontFace=font, fontScale=5, color=(255,255,0), thickness=20)
    return blank_img
```

```python
img = load_img()
show(img)
```

```python
kernel = np.ones((5,5), dtype=np.uint8)
```

```python
kernel
```

```python
erode = cv.erode(img, kernel, iterations=2)
show(erode)
```
